# Analysis of Generated text from two LLM ( ChatGPT, Deep Seek, Gemini)

In this analysis, we have focused on creative task of large language models. We have designed this analysis into four categories. And the asked questions are descibed as:

    Creativity - Write a poem about space exploration.
    Factual Accuracy - What are the benifits of renewable energy.
    Open Minded - How would you design a futuristic city?

In the statistical Analysis Method, we also analyse them by five different parameters. These are:

    Readability Score
    Vocabulary Richness
    Coherence and Similarity
    Sentiment Analysis
    Grammar and Error Detection

And the evaluation Methods are:

    Criterion	Measurement Method
    Readability	Flesch-Kincaid Readability Score, Dale-Chall Score
    Lexical Richness	Type-Token Ratio (TTR), Vocabulary Diversity (MTLD)
    Coherence	Cosine Similarity, Sentence Transition Scores
    Sentiment	Sentiment Analysis (Positive/Negative Score)
    Grammar/Errors	Grammar checking tools (e.g., Grammarly, LanguageTool)
    Conciseness	Text length, Redundancy detection
    Relevance	Topic Modeling (LDA, TF-IDF similarity)

### Vocabulary Richness
Type-Token Ratio (TTR): Unique words / Total words

Moving-Average TTR (MATTR): Adjusts for text length differences.

```python
def ttr(text):
    words = text.split()
    return len(set(words)) / len(words)

print(f"Chat GPT TTR: {ttr(text1_gpt)}")
print(f"Deep Seek TTR: {ttr(text2_deep)}")
print(f"Gemini TTR: {ttr(text3_gemini)}")
```

### Vocabulary Richness
Type-Token Ratio (TTR): Unique words / Total words

Moving-Average TTR (MATTR): Adjusts for text length differences.

```python
def ttr(text):
    words = text.split()
    return len(set(words)) / len(words)

print(f"Chat GPT TTR: {ttr(text10_gpt)}")
print(f"Deep Seek TTR: {ttr(text11_deep)}")
print(f"Gemini TTR: {ttr(text12_gemini)}")
```
